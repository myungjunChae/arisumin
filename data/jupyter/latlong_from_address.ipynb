# 서울시상수도사업본부

```
#아리수
import pandas as pd
import requests

#df = pd.read_csv('/content/arisu.csv', encoding='CP949', names=['address','lat','long'])
df = pd.read_csv('/content/arisu.csv', encoding='CP949')

for index, row in df.iterrows():
  address = row[0]
  
  resp = req(address)
  
  try:
    data = resp.json()
    lat = data['addresses'][0]['y'] #위도
    long = data['addresses'][0]['x'] #경도
    
    df.loc[index, 1] = lat
    df.loc[index, 2] = long
       
  except:
    continue
    
df.to_csv('/content/arisu_latlong.csv')
```

```
#kwater
import pandas as pd
import requests

path = '/content/gdrive/My Drive/'
name = 'kwater.csv'
df = pd.read_csv(path+name)

#drop index
df = df.drop([df.columns[0]], axis='columns') #drop index

#add column
#df['lat'] = 0
#df['long'] = 0

for index, row in df.iterrows():
  address = row[1]
    
  resp = req(address)
  
  try:
    data = resp.json()
    lat = data['addresses'][0]['y'] #위도
    long = data['addresses'][0]['x'] #경도
    
    df.loc[index, 'lat'] = lat
    df.loc[index, 'long'] = long
    
    print(index, address, lat, long)
       
  except:
    continue
    
df.to_csv(path+'kwater_latlong.csv')
```

# JSON Convert

```
df = pd.read_csv(path+'kwater_latlong.csv')
df.columns = ['index','name','address','state','lat','long']

data = df.to_json(orient='records', lines=True, force_ascii=False)
data = data.replace('\n',',')

json = '{{"data":[{data}]}}'.format(data=data)

with open(path + 'kwater.json', 'w') as outfile:
     outfile.write(json)
```
